<h4>Quantum Circuit</h4>

<b>Quantum Circuit is like Digital Circuit in Classical System. We know that a Digital circuit comprises of logic gates (AND-gate,NOT-gate etc) arranged in a way to perform operation on the inputs which are in bits.
So is the Quantum Circuit.A quantum circuit is a sequence of quantum gates to operates on qubits by manipulating their states using quantum gates </b>
<br>
Quantum circuits are built on qubits, which can be in a superposition of 0 and 1 states.
Quantum gates are similar to classical logic gates. These are the common quantum gates:
<p>X (Pauli-X): flips a qubit's state:  If the input qubit is |0‚ü© (represented as [1, 0] in vector form), the X gate turns it into |1‚ü© </p>
<p>H (Hadamard): creates superposition.Makes the quantum system to have all the possible states in existence with their probability</p>
<p>CX (CNOT): entangles qubits.That it connect states together regardless of their separation or distance apart</p>

<h4>Let Us Create a design a Quantum Circuit from a single qubit</h4>

```python
#import QuantumCircuit class from Qiskit
from qiskit import QuantumCircuit


# Create a Quantum Circuit with 1 qubits
qc = QuantumCircuit(1)

# Draw the circuit
qc.draw('mpl')
```

<h4> As you can see, the Circuit has no gate, so there can not be possible operation on the qubits</h4>
Let us repeat the process but with a gate (Hadamard gate)

```python
#import QuantumCircuit class from Qiskit
from qiskit import QuantumCircuit


# Create a Quantum Circuit with 1 qubit
qc = QuantumCircuit(1)

# With Hadamard gate to make the qubit be in superposition
qc.h(0)
# Draw the circuit
qc.draw('mpl')
```

```python
#import QuantumCircuit class from Qiskit
from qiskit import QuantumCircuit


# Create a Quantum Circuit with 2 qubits
qc = QuantumCircuit(2)


# With Hadamard gate to make the first qubit be in superposition
qc.h(0)

# With Hadamard gate to make the second qubit be in superposition
qc.h(1)

# Draw the circuit
qc.draw('mpl')
```

```python
<h4> Let us CX (CNOT) gate:to entangles qubits.</h4>
```

```python
#import QuantumCircuit class from Qiskit
from qiskit import QuantumCircuit


# Create a Quantum Circuit with 2 qubits
qc = QuantumCircuit(2)


# With Hadamard gate to make the first qubit be in superposition
qc.h(0)

# With Hadamard gate to make the second qubit be in superposition
qc.h(1)

qc.cx(0, 1)  # Entangle qubit 0 and qubit 1

# Draw the circuit
qc.draw('mpl')
```
